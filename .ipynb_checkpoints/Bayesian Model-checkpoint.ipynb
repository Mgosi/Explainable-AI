```python
import pandas as pd
import numpy as np
from pgmpy.estimators import K2Score
from pgmpy.models import BayesianModel
from IPython.display import display
from pgmpy.inference import BeliefPropagation
from pgmpy.inference import VariableElimination
from pgmpy.estimators import BayesianEstimator
from pgmpy.estimators import MaximumLikelihoodEstimator
```

```python
import networkx as nx
import pylab as plt
def drawModel(model):
    nx.draw(model, with_labels=True)
```

```python
ds = pd.read_csv("15features.csv")
trainingSeen = pd.read_csv("dataset_seen_training_siamese.csv")
testingSeen = pd.read_csv("dataset_seen_validation_siamese.csv")

ds.iloc[:,1:] -= 1

mergedTemp = trainingSeen.merge(ds, left_on="left" , right_on="imagename")
merged = mergedTemp.merge(ds, left_on="right", right_on = "imagename").iloc[:,3:]

cols = merged.columns.tolist()
c = [cols[1], cols[17]]
c.append(cols[0])
c.extend(cols[2:17])
c.extend(cols[18:])
merged = merged[c]


mergedTestTemp = testingSeen.merge(ds, left_on="left" , right_on="imagename")
mergedTest = mergedTestTemp.merge(ds, left_on="right", right_on = "imagename").iloc[:,3:]

cols = mergedTest.columns.tolist()
c = [cols[1], cols[17]]
c.append(cols[0])
c.extend(cols[2:17])
c.extend(cols[18:])
mergedTest = mergedTest[c]
```

```python
from pgmpy.estimators import HillClimbSearch

hc = HillClimbSearch(ds.iloc[:,1:], scoring_method=K2Score(ds.iloc[:,1:]))
best_model = hc.estimate(max_indegree= 1)
```

```python
drawModel(best_model)
```

```python
best_model.edges()
```

```python
model = BayesianModel([('size_x', 'slantness_x'),
 ('size_x', 'pen_pressure_x'),
 ('size_x', 'letter_spacing_x'),
 ('dimension_x', 'size_x'),
 ('is_continuous_x', 'exit_stroke_d_x'),
 ('is_continuous_x', 'formation_n_x'),
 ('is_continuous_x', 'entry_stroke_a_x'),
 ('slantness_x', 'tilt_x'),
 ('staff_of_a_x', 'is_lowercase_x'),
 ('formation_n_x', 'constancy_x'),
 ('staff_of_d_x', 'is_continuous_x'),
 ('word_formation_x', 'dimension_x'),
 ('constancy_x', 'word_formation_x'),
 ('constancy_x', 'staff_of_a_x'),('size_y', 'slantness_y'),
 ('size_y', 'pen_pressure_y'),
 ('size_y', 'letter_spacing_y'),
 ('dimension_y', 'size_y'),
 ('is_continuous_y', 'exit_stroke_d_y'),
 ('is_continuous_y', 'formation_n_y'),
 ('is_continuous_y', 'entry_stroke_a_y'),
 ('slantness_y', 'tilt_y'),
 ('staff_of_a_y', 'is_lowercase_y'),
 ('formation_n_y', 'constancy_y'),
 ('staff_of_d_y', 'is_continuous_y'),
 ('word_formation_y', 'dimension_y'),
 ('constancy_y', 'word_formation_y'),
 ('constancy_y', 'staff_of_a_y'),('is_lowercase_x','label'),('is_lowercase_y','label'),
                       ('size_x','label'),('size_y','label')])
```

```python
model.fit(merged.iloc[:,2:])
```

```python
drawModel(model)
```

## Inference Method 1

```python
inference = VariableElimination(model)
evid = mergedTest.iloc[:,3:].to_dict(orient = 'records')
query = []
for i in range(len(evid)):
    query.append(inference.map_query(variables = ['label'], evidence = evid[i])['label'])
```

```python
acc = (float)((query == mergedTest['label']).sum()/len(query))*100
print ("Accuracy = " + str(acc))
```

### Inference Method 2

```python
evid = mergedTest.iloc[:,3:].to_dict(orient = 'records')
col = mergedTest.iloc[:,2:].columns.tolist()
varColG = col[16:]
varColG

varColF = col[1:16]
varColF

evidColsF = (mergedTest.iloc[:,2:18]).to_dict(orient = 'records')
```

```python

#belief_propagation.calibrate()

queryG_F0 = []
for evid in evidColsF:
    evid['label'] = 0
    queryG_F0.append(inference.query(variables = varColG, evidence = evid))
```

```python
queryG_F1 = []
for evid in evidColsF:
    evid['label'] = 1
    queryG_F1.append(inference.query(variables = varColG, evidence = evid))
```

```python
queryG_F0[1]['pen_pressure_y'].values
```

```python
featDS = mergedTest.iloc[:,18:]
```

```python
likelihoodRatio = []
predY = []
for i in range(len(queryG_F0)):
    p_gf0 = 1
    p_gf1 = 1
    G_F0 = queryG_F0[i]
    G_F1 = queryG_F1[i]
    for j, feat in enumerate(varColG):
        p_gf0 *= G_F0[feat].values[featDS.iloc[i,j]]
        p_gf1 *= G_F1[feat].values[featDS.iloc[i,j]]
    likelihoodRatio.append(p_gf0/p_gf1)
    #print(p_gf0, p_gf1, likelihoodRatio)
```

```python
predY = []
for i in range(len(likelihoodRatio)):
    if likelihoodRatio[i] > 1.0 :
        predY.append(0)
    else:
        predY.append(1)
acc = (np.asarray(predY)==np.asarray(mergedTest['label'].values.tolist())).sum()/len(predY)*100
print ("Accuracy : ",acc)
```
